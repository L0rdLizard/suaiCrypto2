```python
import matplotlib.pyplot as plt
import numpy as np

def feige_fiat_shamir_false_identification_probability(t, n):
    """
    Рассчитывает вероятность ложной идентификации в протоколе Feige-Fiat-Shamir.

    :param t: количество раундов (итераций)
    :param n: количество бит (или степень защиты)
    :return: вероятность ложной идентификации
    """
    # Вероятность ложной идентификации определяется как (1 / 2^t)
    return 1 / (2 ** t)

# Задаем диапазон значений для количества раундов
t_values = np.arange(1, 21, 1)  # от 1 до 20 раундов
n = 1  # Константа, так как n не влияет на вероятность ложной идентификации в FFS

# Рассчитываем вероятности ложной идентификации для каждого значения t
probabilities = [feige_fiat_shamir_false_identification_probability(t, n) for t in t_values]

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(t_values, probabilities, marker='o', linestyle='-', color='b')
plt.yscale('log')
plt.xlabel('Количество раундов (t)')
plt.ylabel('Вероятность ложной идентификации')
plt.title('Вероятность ложной идентификации в протоколе Feige-Fiat-Shamir')
plt.grid(True, which="both", ls="--")

# Отображаем график
plt.show()
```
