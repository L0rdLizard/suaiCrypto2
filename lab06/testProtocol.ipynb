```python
import matplotlib.pyplot as plt
import numpy as np

def feige_fiat_shamir_false_identification_probability(t, n):
    """
          Feige-Fiat-Shamir.

    :param t:   ()
    :param n:   (  )
    :return:   
    """
    #      (1 / 2^t)
    return 1 / (2 ** t)

#      
t_values = np.arange(1, 21, 1)  #  1  20 
n = 1  # ,   n        FFS

#        t
probabilities = [feige_fiat_shamir_false_identification_probability(t, n) for t in t_values]

#  
plt.figure(figsize=(10, 6))
plt.plot(t_values, probabilities, marker='o', linestyle='-', color='b')
plt.yscale('log')
plt.xlabel('  (t)')
plt.ylabel('  ')
plt.title('     Feige-Fiat-Shamir')
plt.grid(True, which="both", ls="--")

#  
plt.show()
```
